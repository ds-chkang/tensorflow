```python
#Threshold를 바꿔가면서 트래이닝 수행.
#Training 데이터와 Test 데이터를 나누지 않고 실행.
```

```python
import tensorflow.compat.v1 as tf
import matplotlib.pyplot as plt
import numpy as np
tf.disable_v2_behavior()
```

```python
x_data = [[1,2],[2,3],[3,1],[4,3],[5,3],[6,2]]
y_data =[[0],[0],[0],[1],[1],[1]]
```

```python
X = tf.placeholder(tf.float32, shape=[None, 2])
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
W = tf.Variable(tf.random_normal([2, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
H = tf.sigmoid(tf.matmul(X, W) + b)
```

```python
cost = -tf.reduce_mean(Y*tf.log(H)+(1-Y)*tf.log(1-H))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.0001)
```

```python
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
# Threshold를 바꿔가면서 트래이닝 되는 과정 확인.
threshold = 0.0
for i in range(100):
    for step in range(2001):
        cost_val, _ = sess.run([cost, train], feed_dict={X:x_data, Y:y_data})
        #print ("Cost: ", cost_val)
    print ("Training is done.")
    predicted = tf.cast(H > threshold, dtype=tf.float32)
    accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))
    _, _, accuracy_val = sess.run([H, predicted, accuracy], feed_dict={X:x_data, Y:y_data})
    print ("Step: ", step, "  Cost: ", cost_val, "  Accuracy: ", accuracy_val, "  Threshold: ", threshold)        
    threshold += 0.01
```
